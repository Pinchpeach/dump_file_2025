```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import transforms
```

```python
val = np.linspace(0.000001 ,1, 1000)
rn = np.random.normal(0, 1, 10000)

transdata = plt.gca().transData
rotate = transforms.Affine2D().rotate_deg_around(1, 1, 90)

plt.figure(figsize=(12, 12))
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.plot(val, np.sqrt(-2*np.log(val)), color='red',transform = rotate + transdata)
plt.hist(rn, bins=30, density=True, color='blue')
```
