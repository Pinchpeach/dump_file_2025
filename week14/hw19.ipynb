```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func1(x):
    return np.sin(x)**2 * np.cos(x / 4) + 4 * x**4 - 6 * x**3 - 2*x + 6

def func2(x):
    return -np.cos(x/2)**2 / (x**2 + 5) - 2 * np.exp(-(x - 3)**2 / 100)
```

```python
def sumi_fir(x, func, h=1e-5):
    return (func(x + h) - func(x - h)) / (2 * h)
    
def sumi_sec(x, func, h=1e-5):
    return (func(x + 2*h) - 2 * func(x) + func(x - 2*h)) / (4 * h**2)
```

```python
def Line_mini(function, ex=1e-6, a=-100, c=100):
    gr = 0.38197
    b = a + gr * (c - a)
    d = b + gr * (c - b)
    
    while np.abs(c-a) > ex:
        if function(b) < function(d):
            c = d
        else:
            a = b
            
        b = a + gr * (c - a)
        d = b + gr * (c - b)
            
    return b
```

```python
min_func1 = Line_mini(func1)
min_func2 = Line_mini(func2)
print(f"minimum place of function 1: {min_func1} \nminimum place of function 2: {min_func2}")
```

```python
print(f"first derivation of min_func1: {sumi_fir(min_func1, func1)}\nfirst derivation of min_func2: {sumi_fir(min_func2, func2)}")
```

```python
x = np.linspace(-20, 20, 200)
plt.figure(figsize=(10, 5))
plt.yscale('log')
plt.plot(x, func1(x))
plt.scatter(min_func1, func1(min_func1), marker='x', color='red')
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x, func2(x))
plt.scatter(min_func2, func2(min_func2), marker='x', color='red')
```
