## 📌 Matplotlib `figure`, `axes` 객체
`plt.hist()`, `plt.scatter()` 를 단순히 사용해도 간단한 그래프를 그릴수 있지만, 디테일한 설정을 위해서는 figure와 axes 객체에 대한 이해가 필요. 

---
#### **Figure 객체**
- 그래프 전체를 담는 액자의 역할
- 하나의 `Figure` 안에 여러 `axes`를 포함할 수 있다.
- `plt.figure()` 를 사용하여 생성 가능

````python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8, 6))  # fig 변수에 Figure 객체 생성
plt.show() #액자만 있고 그림은 없는 상태
````

#### **Axes 객체**
- 실제로 그래프를 그리는 영역
- 하나의 `Figure` 안에 여러 `axes`를 포함할 수 있다.
- `fig.add_subplot()` 또는 `plt.subplots()`를 사용하여 생성 가능

#### **pyplot example**

````python
def kroupa_imf(m):
    imf = np.zeros(len(m))
    range1 = m < 0.08; range2 = (m >= 0.08) & (m < 0.5); range3 = m >= 0.5
    imf[range1] = m[range1] ** -0.3
    imf[range2] = m[range2] ** -1.3 * (imf[range1][-1]/m[range2][0]**-1.3)
    imf[range3] = m[range3] ** -2.3 * (imf[range2][-1]/m[range3][0]**-2.3)
    return imf, range1, range2, range3

mass = np.linspace(0.01,120,100000)
imf, range1, range2, range3 = kroupa_imf(mass)
````

#### - *Simple usage*

Easy, but redundant and too long

plt.plot() create `figure` and `axes` automaticallly.

````python
plt.plot(mass[range1], imf[range1], label=r'$M<0.08M_{\odot}$')
plt.plot(mass[range2], imf[range2], label=r'$0.08M_{\odot}<M<0.5M_{\odot}$')
plt.plot(mass[range3], imf[range3], label=r'$0.5M_{\odot}<M<120M_{\odot}$')
plt.xlabel(r'$M_{\odot}$', fontdict={'size':14, 'style':'italic'})
plt.ylabel(r'${\xi}(m)$', fontdict={'size':14, 'style':'italic'})
plt.xscale('log'); plt.yscale('log')
plt.title('Kroupa IMF')
plt.legend()
````

#### - *Combining with for loops*

Better visibility, but how to handle the labels?

````python
for range in [range1, range2, range3]:
    plt.plot(mass[range], imf[range])
plt.xlabel(r'$M_{\odot}$', fontdict={'size':14, 'style':'italic'})
plt.ylabel(r'${\xi}(m)$', fontdict={'size':14, 'style':'italic'})
plt.xscale('log'); plt.yscale('log')
plt.title('Kroupa IMF')
````

````python
ranges = [range1, range2, range3]
labels = [r'$M<0.08M_{\odot}$', r'$0.08M_{\odot}<M<0.5M_{\odot}$', r'$0.5M_{\odot}<M<120M_{\odot}$']
````

````python
for i in np.arange(3):
    plt.plot(mass[ranges[i]], imf[ranges[i]], label=labels[i])
plt.xlabel(r'$M_{\odot}$', fontdict={'size':14, 'style':'italic'})
plt.ylabel(r'${\xi}(m)$', fontdict={'size':14, 'style':'italic'})
plt.xscale('log'); plt.yscale('log')
plt.title('Kroupa IMF')
plt.legend()
````

````python
for range, label in zip(ranges, labels):
    plt.plot(mass[range], imf[range], label=label)
plt.xlabel(r'$M_{\odot}$', fontdict={'size':14, 'style':'italic'})
plt.ylabel(r'${\xi}(m)$', fontdict={'size':14, 'style':'italic'})
plt.xscale('log'); plt.yscale('log')
plt.title('Kroupa IMF')
plt.legend()
````

각 mass regime에 대해 IMF를 따로 그리고 싶으면?

````python
for range, label in zip(ranges, labels):
    plt.plot(mass[range], imf[range], label=label)
    plt.xlabel(r'$M_{\odot}$', fontdict={'size':14, 'style':'italic'})
    plt.ylabel(r'${\xi}(m)$', fontdict={'size':14, 'style':'italic'})
    plt.xlim(0.01,120); plt.ylim(10**-3, 10)
    plt.xscale('log'); plt.yscale('log')
    plt.title('Kroupa IMF')
    plt.legend()
    plt.show()
````

#### - *객체 지향 방식*

`Figure`와 `Axes`를 직접 생성하고 조작하여 보다 명확하게 제어가 가능.

- `.add_subplot()` method 활용

````python
fig = plt.figure(figsize=(12, 3)) # fig 변수에 Figure 객체 생성  
ax1 = fig.add_subplot(1,4,1)  # 변수명 ax1에 axes 객체 생성 # (1,4,1) -> 1행, 4열, 에서 1번 위치치
ax1.plot(mass[range1], imf[range1])
ax1.set_xlim(0.01,120); ax1.set_ylim(10**-3, 10); ax1.set_xscale('log'); ax1.set_yscale('log')
ax2 = fig.add_subplot(1,4,2)
ax2.plot(mass[range2], imf[range2])
ax2.set_xlim(0.01,120); ax2.set_ylim(10**-3, 10); ax2.set_xscale('log'); ax2.set_yscale('log')
ax3 = fig.add_subplot(1,4,3)
ax3.plot(mass[range3], imf[range3])
ax3.set_xlim(0.01,120); ax3.set_ylim(10**-3, 10); ax3.set_xscale('log'); ax3.set_yscale('log')
ax4 = fig.add_subplot(1,4,4)
ax4.plot(mass, imf)
ax4.set_xlim(0.01,120); ax4.set_ylim(10**-3, 10); ax4.set_xscale('log'); ax4.set_yscale('log')
plt.tight_layout() # axes간 간격 조정, 사용시 심미성이 좋아지는 경우가 많음
````
