```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x, y = np.arange(100)-50, np.arange(100)-50
X, Y = np.meshgrid(x,y)
r = np.sqrt(X**2+Y**2)
plt.imshow(r)
plt.xlabel('X', fontsize=14); plt.ylabel('Y', fontsize=14)
cbar = plt.colorbar()
cbar.set_label('r', fontsize=14)
plt.show()
```

```python
# r<20 인 영역의 데이터만 얻고싶다면?
idx = r<20
print(r[idx].shape)
plt.imshow(r[idx]) 
```

```python
#np.where 활용
r_new = np.where(idx, r, np.nan)
plt.imshow(r_new)
plt.xlabel('X', fontsize=14); plt.ylabel('Y', fontsize=14)
cbar = plt.colorbar()
cbar.set_label('r', fontsize=14)
plt.show()
```
