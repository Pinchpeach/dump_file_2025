```python
import homework2 as shuf
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Box_Muller_tf():
    x1 = shuf.myran()
    x2 = shuf.myran()
    
    t = np.sqrt(-2*np.log(x1))
    
    y1 = t*np.cos(2*np.pi*x2)
    y2 = t*np.sin(2*np.pi*x2)
    
    return y1, y2
```

```python
mean = 10
std = 2
n = 10**4

result = np.zeros(n)
for _ in range(n//2):
    y1, y2 = Box_Muller_tf()
    result[2*_] = mean + std*y1
    result[2*_+1] = mean + std*y2
    
result_np = np.random.normal(mean, std, n)
```

```python
fig = plt.figure(figsize=(12, 6))
fig1 = fig.add_subplot(1, 2, 1)
cmp1 = fig1.hist(result, bins=50, density=True, edgecolor='black')
fig2 = fig.add_subplot(1, 2, 2)
cmp2 = fig2.hist(result_np, bins=50, density=True, edgecolor='black')
plt.show()
```

```python
# codes for compare

cal = cmp1[0] - cmp2[0]
cal = cal**2
MSE = np.sum(cal) / len(cal)
print("MSE is", MSE)
```
