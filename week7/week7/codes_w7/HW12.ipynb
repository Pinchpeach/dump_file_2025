$ f(x) = \begin{cases}
& 1, \ (0 < x < \pi) \\ & -1, \ (-\pi < x < 0)
\end{cases} , \ \ \ f(x + 2\pi) = f(x) $

$ f(x) = {1 \over {2\pi}}\int_{-\pi}^{\pi}f(x)dx + \sum_{n = 1}^{\inf}({1 \over {\pi}}\int_{-\pi}^{\pi}f(x)\cos{nx}dx\cos{nx} \ + \ {1 \over {\pi}}\int_{-\pi}^{\pi}f(x)\sin{nx}dx \sin{nx} ) $

by given condition
$ \\ {1 \over {2\pi}}\int_{-\pi}^{\pi}f(x)dx \\ $
$ = {1 \over {2\pi}}(\int_{-\pi}^{0}-1dx + \int_{0}^{\pi}1dx) = 0 $

$ \\ \\ \\ $

$ \\ {1 \over {\pi}}\int_{-\pi}^{\pi}f(x)\cos{nx}dx\cos{nx} \\
 = {1 \over {\pi}}(\int_{-\pi}^{0}-1\cos{nx}dx + \int_{0}^{\pi}1\cos{nx}dx) \\
 = {1 \over {n\pi}}(-\sin{0} + \sin({-n\pi}) + \sin({n\pi}) -\sin{0}) \\ $
 ( n is integer )
$ \\ = 0$

$ \\ \\ \\ $ 

$ {1 \over {\pi}}\int_{-\pi}^{\pi}f(x)\sin{nx}dx \\
= {1 \over {\pi}}(\int_{-\pi}^{0}-\sin{nx}dx + \int_{0}^{\pi}\sin{nx}dx) \\
= {1 \over {n\pi}}(\cos({0}) - \cos({-n\pi}) + \cos({0}) - \cos({n\pi})) \\ 
= {2 \over {n\pi}}(1 - \cos({n\pi}))$

$ f(x) = \sum_{n = 1}^{\inf}({2 \over {n\pi}}(1 - \cos({n\pi}))\sin({nx})) \\ $
if n is odd
$ \\ \\
f(x) = {4 \over {n\pi}}\sin({nx}) \\ \\$
$ \\ $ 
if n is even
$ \\ \\
f(x) = 0$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def given_func(x):
    if (x > 0 and x < np.pi):
        return 1
    elif (x < 0 and x > -np.pi):
        return -1
    else:
        return 0
    
func_vec = np.vectorize(given_func)
```

```python
gap = np.linspace(-np.pi, np.pi, 1000)
result_list = []
for _ in range(200):
    n = 2 * _ + 1
    result_list.append(_)
    result_list[_] = (4 * np.sin(n * gap) / (n * np.pi))
```

```python
approx = np.zeros(len(gap))
rmse_arr = np.zeros(200)

plt.figure()
plt.plot(gap, func_vec(gap), color='black')
for _ in range(200):
    approx += result_list[_]
    rmse_arr[_] = np.sqrt(np.mean((func_vec(gap) - approx) ** 2))
    plt.plot(gap, approx, alpha=0.5)
```

```python
plt.figure()
for _ in range(len(rmse_arr)):
    print("%3dth RMSE: %.7f" %(_ + 1, rmse_arr[_]))
plt.plot(rmse_arr)
```
